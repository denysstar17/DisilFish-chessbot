```python
import re
import pandas as pd

game_pattern = r'(\[Event.*?\])\s*\n\n(1\..*?)\n\n'

all_games = re.findall(game_pattern, pgn_string, re.DOTALL)

info_pattern = r'\[Date "(.*?)"\]\s*\n\[Round "(.*?)"\]\s*\n\[White "(.*?)"\]\s*\n\[Black "(.*?)"\]\s*\n\[Result "(.*?)"\]'

games_data = []

for info, moves in all_games:
    match = re.search(info_pattern, info)
    if match:
        date = match.group(1)
        rnd = match.group(2)
        white_player = match.group(3)
        black_player = match.group(4)
        result = match.group(5)
        games_data.append({
            'White': white_player,
            'Black': black_player,
            'Result': result,
            'Date': date,
            'Game': moves.strip()
        })

df = pd.DataFrame(games_data)
```
